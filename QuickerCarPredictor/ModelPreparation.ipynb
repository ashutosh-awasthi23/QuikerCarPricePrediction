```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
import pickle
```

```python
Car=pd.read_csv("quikr_car.csv")
Car
```

```python
df=Car.copy()
df
```

```python
df.columns
```

```python
df.rename(columns=
    {
        "kms_driven":"km",
        "fuel_type":"fuel"
    },
    inplace=True
)
df.columns
```

```python
df.info()
```

#### Dealing with Kilometers

```python
df["km"]
```

```python
#km columns has km in it
df["km"]=df["km"].str.split().str.get(0)
df["km"]=df["km"].str.replace(",","")
df["km"]=pd.to_numeric(df["km"],errors="coerce")

```

#### Dealing with year columns

```python
df["year"]=pd.to_numeric(df["year"],errors="coerce")
df.info()
```

#### Dealing with price

```python

df["Price"]=df["Price"].str.replace(",","")
df["Price"]=pd.to_numeric(df["Price"],errors="coerce")
df
```

```python
df["Price"].unique()
```

#### Dealing with name


```python
df["name"]=df["name"].str.split(" ").str.slice(0,3).str.join(" ")
```

#### Outlier Removal

```python
q1=df["Price"].quantile(0.25)
q3=df["Price"].quantile(0.75)
iqr=q3-q1
mini=q1-1.5*iqr
maxi=q3+1.5*iqr
df=df[df["Price"]<=maxi]
```

#### Drop na values

```python
df.dropna(inplace=True)
df["year"]=df["year"].astype("int64")
```

```python
df.info()
```

#### Saving clean data to csv

```python
df
```

```python
df.to_csv("clean_quiker.csv")
```

#### Splitting I/O columns

```python
y=df["Price"]
X=df.drop("Price",axis=1)
```

#### Column Transformation

```python
ohe=OneHotEncoder()
ohe.fit(X[["name","fuel","company"]])
```

```python
ct= make_column_transformer((OneHotEncoder(handle_unknown='ignore'),["name","fuel","company"]),remainder="passthrough")
```

#### Finding out correct random state for better accuracy

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=Pipeline([("CT",ct),("LR",lr)])
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

#### Model Taining

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=Pipeline([("CT",ct),("LR",lr)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```

#### Dumping Model

```python
pickle.dump(pipe,open('LinearRegressionModel.pkl','wb'))
```
